# Statistical Analysis

### Chi-Squared Test

We can apply Chi-Squared Test to check is there is any association between the "Country" of the production and the "Category"(TV show or movie)

```python
netflix_df.head(5)
```

```python
netflix_df_stats = netflix_df # storing it in another variable for safe practice
```

```python
netflix_df_stats.head()
```

```python
from scipy.stats import chi2_contingency

# Split the 'Country' column and expand it to create a row for each country
expanded_df = netflix_df_stats.assign(Country=netflix_df_stats['Country'].str.split(',')).explode('Country').reset_index(drop=True)

# Remove any leading or trailing whitespace from the 'Country' column
expanded_df['Country'] = expanded_df['Country'].str.strip()

# Create a contingency table for all countries and the 'Category' column
all_countries_contingency_table = pd.crosstab(expanded_df['Country'], expanded_df['Category'])

# Perform the Chi-squared test
chi2_stat_all, p_val_all, dof_all, expected_all = chi2_contingency(all_countries_contingency_table)
```

```python
chi2_stat_all
```

```python
p_val_all
```

### Conclusions

I performed a Chi-squared test to assess the association between the countries (considering all countries listed for co-productions) and the content category (TV show or movie).

The results are:

Chi-squared statistic: 1089.7635035197538
 
p-value: 1.8115436927008787e-157

 
Interpreting the results:

Chi-squared statistic: The value of 1089.76 suggests a significant difference between the observed counts in the contingency table and what would be expected if country of production and content category were independent. A larger chi-squared value indicates a greater discrepancy between observed and expected counts.

p-value: The p-value measures the evidence against the null hypothesis. A very small p-value, like the one you obtained, indicates strong evidence against the null hypothesis.

Given the extremely small p-value, we can reject the null hypothesis. This means there is a statistically significant association between the country of production and the content category (TV show or movie) in the dataset.

Conclusion: The type of content (whether it's a TV show or a movie) on Netflix is not distributed independently of its country of production. Some countries may predominantly produce more TV shows, while others might produce more movies, or there could be other patterns of association.

# Visualisations

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the visualizations
sns.set_style("whitegrid")

# Initialize the figure
plt.figure(figsize=(10, 6))

# Plotting the distribution of Categories (Movies vs. TV Shows)
sns.countplot(data=netflix_df, x='Category', palette='pastel')
plt.title('Distribution of Categories on Netflix', fontsize=15)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

The visualization shows the distribution of content categories on Netflix. It's evident that there are more movies than TV shows available on the platform.

```python
# Initialize the figure
plt.figure(figsize=(14, 7))

# Plotting the distribution of Ratings
sns.countplot(data=netflix_df, x='Rating', order=netflix_df['Rating'].value_counts().index, palette='pastel')
plt.title('Distribution of Ratings on Netflix', fontsize=15)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

The visualization provides an overview of content ratings on Netflix. It appears that the majority of content is rated as "TV-MA" (Mature Audiences), followed by "TV-14" (Parents Strongly Cautioned) and "TV-PG" (Parental Guidance).

```python
# Extracting the release year from the Release_Date column
netflix_df['Release_Year'] = pd.to_datetime(netflix_df['Release_Date'], errors='coerce').dt.year

# Initialize the figure
plt.figure(figsize=(18, 7))

# Plotting the distribution of Release Years
sns.countplot(data=netflix_df, x='Release_Year', palette='pastel', order=netflix_df['Release_Year'].value_counts().index)
plt.title('Distribution of Release Years on Netflix', fontsize=15)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

The visualization displays the distribution of release years for content available on Netflix. We can observe a few key points:

There's a significant increase in content released on Netflix in recent years, especially from around 2016 onwards.
The year 2020 seems to have the highest count of content released, followed by 2019 and 2021.
